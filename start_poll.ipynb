```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')
```

```python
students = {
        'Валентина Козина': 'yes',
        'Никита Трибунцев': 'yes',
        'Aleksey Zabelin': 'yes',
    
        'Настя': 'maybe',
        'Alexandr Sterkhov': 'maybe',
        'Vilena B.': 'maybe',
        'Господин Инженер': 'maybe',
        
        'Алина Пуциньш': 'later',
        'Chingis Mironov': 'later',
}

columns = ['ready_to_start']

print(len(students))
```

```python
project_pool = pd.DataFrame.from_dict(data=students, columns=columns, orient='index')
project_pool.reset_index(drop=False, inplace=True)
project_pool.rename(columns={'index': 'sudent_name'}, inplace=True)
# display(project_pool)
```

```python
sizes = project_pool.groupby(by='ready_to_start')['sudent_name'].count()
labels = sizes.index
```

```python
colors = sns.color_palette("Greens")

fig, ax = plt.subplots(figsize=(5, 5)) # a figure with a single Axes
ax.pie(sizes, labels=labels, autopct='%1.0f%%',
        startangle=137, colors=colors,
        wedgeprops={"linewidth": 1, "edgecolor": "white"},
        textprops={'color': 'forestgreen', 'fontsize': 13})

ax.set_xlabel('Готовы ли начать практику на реальных данных?',
              fontsize=14, family='sans-serif', color='forestgreen')
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
